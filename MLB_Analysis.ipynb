
## Analysis of Career Longevity and Performance in Baseball Pitching

### Introduction

The dataset covers a range of players, providing a comprehensive view of pitching performance across different career spans. However, it's important to note that the data represents cumulative career statistics, which limits our ability to analyze year-over-year changes but still allows for insightful comparisons between players of varying career lengths.

## Step 1: Data Preprocessing

The initial step in our analysis involved preparing the dataset for exploration and analysis. This preparation, known as data preprocessing, included the following key actions:

### Handling Missing Values

Given the nature of our dataset and the importance of having complete records for accurate analysis, we opted to remove any rows containing missing values. This approach ensures that our analysis is based on complete and reliable data, although it may slightly reduce the total number of records available for examination.

Here's the code used for this step:

```python
# Dropping rows with any missing values
pitching_data_cleaned = pitching_data.dropna()
```

### Converting Data Types

Two columns in our dataset, 'Save Opportunity' and 'AVG', were identified as having incorrect data types for the kind of analysis we intend to perform. Specifically, these columns were initially read as object types (likely strings), but they represent numerical values. Therefore, we converted these columns to floats. During this conversion, any non-numeric values were turned into NaN, which were then subsequently removed from the dataset.

Here's the code for the conversion:

```python
# Convert 'Save Opportunity' and 'AVG' to float and handle non-convertible values
pitching_data_cleaned['Save Opportunity'] = pd.to_numeric(pitching_data_cleaned['Save Opportunity'], errors='coerce')
pitching_data_cleaned['AVG'] = pd.to_numeric(pitching_data_cleaned['AVG'], errors='coerce')

# Drop any rows that resulted in NaN after conversion
pitching_data_cleaned = pitching_data_cleaned.dropna()
```

### Final Dataset Check

After these preprocessing steps, we conducted a final check to ensure that there were no missing values and that all data types were appropriate for our analysis. This check confirmed that our dataset was now clean and ready for the next stages of our study.

````python
# Combined code for data preprocessing

# Import necessary library
import pandas as pd
import numpy as np

# Load the data from the CSV file
file_path = 'baseball_pitcher.csv'  # Replace with your file path
pitching_data = pd.read_csv(file_path)

# Step 1: Dropping rows with any missing values
pitching_data_cleaned = pitching_data.dropna()

# Step 2: Convert 'Save Opportunity' and 'AVG' to float and handle non-convertible values
pitching_data_cleaned['Save Opportunity'] = pd.to_numeric(pitching_data_cleaned['Save Opportunity'], errors='coerce')
pitching_data_cleaned['AVG'] = pd.to_numeric(pitching_data_cleaned['AVG'], errors='coerce')

# Step 3: Drop any rows that resulted in NaN after conversion
pitching_data_cleaned = pitching_data_cleaned.dropna()

# Final check of the cleaned data
print("Final Missing Values:\n", pitching_data_cleaned.isnull().sum())
print("\nFinal Data Types:\n", pitching_data_cleaned.dtypes)
print("\nPreview of Cleaned Data:\n", pitching_data_cleaned.head())
````

## Step 2: Grouping Players by Career Length

With the data cleaned and prepared, the next step was to group the players based on their career lengths. Career length is a significant factor in analyzing a player's performance, as it often correlates with experience, resilience, and changes in playing style.

### Determining Career Length

In the absence of explicit career duration data, we used the 'Games Played' metric as a proxy to infer the length of a player's career. This choice is grounded in the assumption that the number of games played is a reasonable indicator of a player's experience and longevity in the sport.

### Categorizing Players into Groups

To categorize the players, we first examined the distribution of the 'Games Played' using a histogram. The distribution showed a clear right-skew, indicating that most players have a moderate number of games played, with fewer players having very long careers.

Based on the histogram, we defined bins that represent different career length groups. These bins were chosen to capture a range of career stages, from newcomers to veterans:

- `0-100`: Newcomers or players with brief careers
- `101-200`: Players with short to moderate-length careers
- `201-300`, `301-400`, ...: Successively longer career lengths
- `901-1000`: Players with the longest careers

### Grouping Code

Using pandas' `cut` function, we mapped each player to a corresponding career length group based on the number of games they played:

```python
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800', '801-900', '901-1000']
pitching_data_cleaned['Career Length Group'] = pd.cut(pitching_data_cleaned['Games Played'], bins=bins, labels=labels)
```

### Observations from Grouping

The grouping revealed that the majority of players have career lengths in the `101-200` and `201-300` games played range, suggesting that these are the most common career lengths. Few players have careers extending beyond 900 games, highlighting the rarity of extremely long careers in professional baseball.

With these groups established, we can now proceed to analyze the data to uncover trends and insights related to career longevity and its impact on a player's performance metrics.


````python
import matplotlib.pyplot as plt

# Analyze the distribution of 'Games Played' or 'Games Started'
plt.hist(pitching_data_cleaned['Games played'], bins=30, edgecolor='black')
plt.title('Distribution of Games Played')
plt.xlabel('Games Played')
plt.ylabel('Number of Players')
plt.show()
````

````python
# Define the bins for grouping
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

# Create a new column 'Career Length Group' based on these bins
pitching_data_cleaned['Career Length Group'] = pd.cut(pitching_data_cleaned['Games played'], bins, labels=['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800', '801-900', '901-1000'])

# Check the distribution of players in each group
group_distribution = pitching_data_cleaned['Career Length Group'].value_counts()
print(group_distribution)
````

## Step 3: Statistical Analysis and Visualization

Our statistical analysis aims to explore the relationship between career length and various pitching performance metrics. It is important to note that some metrics, such as total strikeouts and wins, are cumulative and thus tend to increase with the number of games played. This inherent correlation means that pitchers with longer careers will naturally accumulate higher totals.

### Cumulative vs. Rate Statistics

While cumulative statistics offer value, they can be misleading when comparing players with varying lengths of careers. To address this, analysts often use rate statistics such as strikeouts per nine innings (K/9) or win-loss percentage, which normalize performance metrics and allow for fairer comparisons. However, for the purpose of initial visualization and exploration, we have chosen to present the cumulative statistics as they are, with a plan to delve into rate-based analysis in subsequent steps.

### Visualizations

We have created visual representations for four key performance metrics: strikeouts, WHIP, wins, and ERA. These visuals serve as an initial step to identify any apparent trends across different career length groups. We have used the mean value within each group for these metrics, which, despite the limitations of cumulative statistics, can still provide some insight into the relationship between career length and performance.


### Interpretation

In interpreting these visuals, we will consider the nature of each statistic. For cumulative totals like strikeouts and wins, we will be cautious not to draw direct conclusions about a player's skill level based on these numbers alone. Instead, we will use these figures as a stepping stone to deeper analysis, incorporating rate statistics for a more nuanced understanding of a pitcher's performance relative to their career length.

Future analyses will focus on such rate statistics to mitigate the effects of career length and provide a more level playing field for comparing players. This will allow us to more accurately assess the impact of career longevity on performance efficiency.

````python
# Calculate descriptive statistics for each group
grouped_data = pitching_data_cleaned.groupby('Career Length Group')
descriptive_stats = grouped_data[['Earned run Average', 'Strikeouts', 'WHIP', 'Win']].agg(['mean', 'median', 'std']).reset_index()
````

## Step 4: Correlation Analysis Between Career Length and Performance Metrics

In this part of our analysis, we seek to quantify the relationships between career length and various pitching performance metrics using correlation analysis. This will provide us with insights into how these metrics tend to change with the number of games played by a pitcher.

### Computing Correlation Coefficients

We computed Pearson correlation coefficients to examine linear relationships between 'Games Played' and key performance metrics such as ERA, strikeouts, WHIP, and wins. It is crucial to note that correlation does not imply causation, and these coefficients simply measure the strength and direction of linear relationships.

### Visualizing Correlations

We have visualized the correlation matrix using a heatmap, which allows us to easily identify and interpret these relationships.

Positive Correlation: A positive correlation coefficient indicates that as career length increases, the metric tends to increase as well.

Negative Correlation: A negative correlation coefficient suggests that as career length increases, the metric tends to decrease.

No Correlation: A coefficient near zero suggests no linear relationship between career length and the metric.


### Interpreting Correlations

Our interpretation of the heatmap will focus on understanding how the experience gained over a career impacts measurable performance. We anticipate seeing positive correlations for some metrics, such as total wins and strikeouts, given their cumulative nature. However, we are particularly interested in rate-based metrics like ERA and WHIP, which could reveal how efficiency and effectiveness might evolve with career progression.



````python
# Calculate the Pearson correlation coefficient for each metric with respect to 'Games Played'
correlation_metrics = pitching_data_cleaned[['Games played', 'Earned run Average', 'Strikeouts', 'WHIP', 'Win']].corr(method='pearson')

# Focus on correlations with 'Games Played'
career_length_correlation = correlation_metrics['Games played'].drop('Games played')  # drop self-correlation
print(career_length_correlation)
````

````python
sns.heatmap(correlation_metrics, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Pitching Metrics')
plt.show()
````

## Step 5: Peak Performance Analysis

In this step, we aimed to uncover at which stages in their careers pitchers typically exhibit peak performance. This analysis is particularly insightful as it goes beyond mere cumulative statistics to explore how efficiency and effectiveness vary over the course of a pitcher's career.

### Methodology

We categorized players into different career length groups based on the number of games played. The key metrics analyzed included ERA, WHIP, strikeouts, and wins. Our goal was to identify the career length group that showed the best average performance for each of these metrics.

### Calculating and Visualizing Average Performance Metrics

For each career length group, we calculated the average of key performance metrics. These averages were then visualized to provide a clear comparison across groups.

### Identifying the Peak Performance Group

We analyzed the data to determine which career length group typically shows peak performance for each metric. This was identified by:

- The lowest average values for metrics like ERA and WHIP, where lower is better.
- The highest average values for metrics like strikeouts and wins, where higher is generally better.

### Observations and Interpretation

Our analysis revealed that:

- Pitchers in the '801-900 games' group showed the lowest ERA and WHIP, indicating their peak efficiency in preventing runs. 
- These findings suggest that pitchers' skills and performance efficiency evolve in complex ways throughout their careers. The identified peak performance groups provide insights into the typical career trajectory of successful pitchers in baseball.

This step in our analysis offered valuable insights into the career progression of pitchers, challenging the conventional understanding of experience and performance in sports.

---


````python
# Calculate the mean of key performance metrics for each career length group
average_metrics_by_group = pitching_data_cleaned.groupby('Career Length Group')[['Earned run Average', 'WHIP', 'Strikeouts', 'Win']].mean()

# Display the average metrics for each group
print(average_metrics_by_group)
````

````python
# Visualizing the average performance metrics for each career length group
average_metrics_by_group.plot(kind='bar', subplots=True, layout=(2,2), figsize=(12, 10), legend=True)
plt.tight_layout()
plt.show()
````

````python
# Define whether lower or higher values are better for each metric
lower_is_better = {
    'Earned run Average': True,
    'WHIP': True,
    'Strikeouts': False,
    'Win': False
}

# Identifying the group with the best performance for each metric
peak_performance_group = {}
for metric, lower_better in lower_is_better.items():
    if lower_better:
        peak_performance_group[metric] = average_metrics_by_group[metric].idxmin()
    else:
        peak_performance_group[metric] = average_metrics_by_group[metric].idxmax()

# Print the peak performance group for each metric
print(peak_performance_group)
````
